```python
from resources.game import Game
from resources.breadth_first_solver import BreadthFirstSolver
```

```python
class ElectomagneticMoat(Game):
    def __init__(self, puzzle_pieces):
        self.puzzle_pieces = puzzle_pieces
        
    @property
    def initial_state(self):
        """The initial state of the game."""
        return (0, ())
    
    def next_moves(self, state):
        # TODO - Potentially need to check for duplicates
        last_number = state[0]
        remaining_pieces = [p for p in self.puzzle_pieces if p not in state[1]]         
        return (
            p for p in remaining_pieces
            if p[0] == last_number or p[1] == last_number
        )
    
    def next_state(self, state, move):
        last_number = move[0] if state[0] == move[1] else move[1]
        return (last_number, state[1] + (move,))
    
    def is_end_state(self, state):
        return False
    
    def score(self, state):
        return sum(a + b for (a, b) in state[1])
 
```

```python
moat = ElectomagneticMoat([
    (0, 2),
    (2, 2),
    (2, 3),
    (3, 4),
    (3, 5),
    (0, 1),
    (10, 1),
    (9, 10),
])
solver = BreadthFirstSolver()
solver.solve(moat)
```
